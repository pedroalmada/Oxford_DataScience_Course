```python
#this to set differnt names to the objects with the help of self. 
class Student:  
    def set_name (self,name): #self can be substituted for any object of the class #in Python self is automatically passed 
        self.name=name

s1=Student() 
s1.set_name('John')
print(s1.name)

s2=Student() 
s2.set_name('Sam')
print(s2.name)

```
