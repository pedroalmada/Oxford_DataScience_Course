```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 10, 1)
x
```

```python
y = x**2
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('A graph')
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(2, 2, 4)
plt.plot(x, y, 'b')
```

```python
fig = plt.figure(figsize=(5,5))
axes = fig.add_axes([0.5, 0.5, 0.6, 0.6])
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=6, figsize=(4,4))
#plt.tight_layout()
for ax in axes:
    ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y, color = 'red', linewidth = 30, alpha = 0.1)
plt.grid()
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y, color = 'red', linewidth = 5, linestyle = '--')
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(x,y, color = 'red', linewidth = 1, marker = '3')
```

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12])
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77]
plt.scatter(x, y)
```

```python
a, b = np.polyfit(x, y, 1)
plt.scatter(x, y)
plt.plot(x, a*x + b)
```

```python
a, b, c = np.polyfit(x, y, 2)
plt.scatter(x, y)
plt.plot(x, a*x^2 + b*x + c)
```
