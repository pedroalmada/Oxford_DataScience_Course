```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.get_dataset_names()
```

```python
x = np.arange(1, 10, 1)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77]
plt.scatter(x, y)

x = [2, 2, 8, 1, 15, 8, 12, 9, 7, 3, 11, 4, 7]
y = [100, 105, 84, 105, 90, 99, 90, 95, 94, 100, 79, 112, 91]
plt.scatter(x, y)

a, b = np.polyfit(x, y, 1)
plt.plot(x, a*np.array(x) + b)
```

```python
x = np.arange(1, 10, 1)
y = x**2
sns.scatterplot(x=x, y=y)
```

```python
x1 = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12]
y1 = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77]

x2 = [2, 2, 8, 1, 15, 8, 12, 9, 7, 3, 11, 4, 7]
y2 = [100, 105, 84, 105, 90, 99, 90, 95, 94, 100, 79, 112, 91]

sns.regplot(x=x1, y=y1, label='Group 1')
sns.regplot(x=x2, y=y2, label='Group 2')
plt.legend()
```
