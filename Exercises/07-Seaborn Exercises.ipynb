```python
# Create the scatter plot with marginal histograms
sns.jointplot(
    x='fare', 
    y='age', 
    data=titanic, 
    kind='scatter',  # Creates scatter plot
    marginal_kws={'bins': 30}  # Sets the number of bins in marginal histograms
)

# Show the plot
plt.show()
```

```python
# Create the histogram for 'fare'
sns.histplot(data=titanic, x='fare', color='red', kde=False, bins=30, alpha=0.6)

# Add labels
plt.xlabel('fare')
plt.ylabel('count')

# Show the plot
plt.show()
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# Create a box plot for age across the different classes
sns.boxplot(x='class', y='age', data=titanic, palette='pastel')

# Labeling the axes for clarity
plt.xlabel("Class")
plt.ylabel("Age")

# Show the plot
plt.show()
```

Explanation:
sns.boxplot():

x='class': This defines the classes (First, Second, Third) along the x-axis.
y='age': This defines the age distribution for each class on the y-axis.
data=titanic: Specifies the dataset to use for the plot.
palette='pastel': Gives the plot the soft pastel color scheme shown in the image.
plt.xlabel() and plt.ylabel():

These label the axes to match the plot shown.
plt.show():

Ensures the plot is displayed in your notebook or environment.

```python
# Select only numeric columns from the Titanic dataset
numeric_titanic = titanic.select_dtypes(include=['number'])

# Plot the heatmap for the correlation matrix of numeric columns
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_titanic.corr(), annot=True, cmap='coolwarm', cbar=True)
plt.title('titanic.corr()')
plt.show()
```
